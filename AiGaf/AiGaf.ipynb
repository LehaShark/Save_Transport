```python
def do_spec (save_path, filename):

    fs = 44100
    sig, fs = librosa.load(folder + filename + '.wav', sr=None)
    pylab.figure(figsize = (1, 1))
    pylab.axis('off') # no axis
    pylab.axes([0., 0., 1., 1.], frameon=False, xticks=[], yticks=[]) # Remove the white edge
    S = librosa.feature.melspectrogram(y=sig, sr=fs)
    librosa.display.specshow(librosa.power_to_db(S, ref=np.max))
    pylab.savefig(save_path + filename, bbox_inches=None, pad_inches=0, dpi = 150)
    pylab.close()
```

```python
def get_img_from_fig(fig):
    
    buf = io.BytesIO()
    fig.savefig(buf, format='png', bbox_inches=None, pad_inches=0, dpi = 150)
    buf.seek(0)
    img_arr = np.frombuffer(buf.getvalue(), dtype=np.uint8)
    buf.close()
    img = cv2.imdecode(img_arr, 1)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    return img
```
